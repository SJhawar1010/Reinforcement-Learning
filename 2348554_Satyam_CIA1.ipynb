```python
import numpy as np
import random

eps = 0.1
nround = 100
nmach = 3

rewards = [0] * nmach
plays = [0] * nmach
erewards = [0.0] * nmach

tprob = [0.3, 0.5, 0.7]
for i in range(nround):
    if random.random() < eps:
        machine = random.randint(0, nmach - 1)
    else:
        machine = np.argmax(erewards)

    reward = 1 if random.random() < tprob[machine] else 0

    plays[machine] += 1
    rewards[machine] += reward

    erewards[machine] = rewards[machine] / plays[machine]

    print(f"Round {i+1}: Machine {machine + 1} was chosen, reward = {reward}")
    print(f"Estimated rewards: {erewards}")
    print(f"Plays: {plays}\n")

print("Final estimated rewards:", erewards)
print("Total plays per machine:", plays)
```
